```python
#States are defined as M,C,B
#M=Count of missionary
#C=Count of cannibal
#B=Bank of river(1-start side, 0-goal side)

start =[3,3,1]
goal=[0,0,0]
states=[]
for i in range(4):
    for j in range(4):
        for k in range(2):
            states.append([i,j,k])

#a non achievable state            
states.remove([0,0,1])

queue=[]
visited=[]
path=[]


#checks whether a state is valid or not
def validState(state):
    if state in states:
        if((state[1]>state[0]) and (state[0]!=0)):
            return False
        elif(((3-state[1])>(3-state[0])) and ((3-state[0])!=0)):
            return False
        else:
            return True

#get possible next states(child) of a state
def nextStates(state):
    childstates=[]
    
    #switch the bank/river side
    if(state[2]==0):
        b_new=1
    if(state[2]==1):
        b_new=0
    
    for i in range(0,3):
        for j in range(0,3):
            if(i+j<=2 and i+j!=0):
                
                #if reaching start side, add values, if leaving, reduce values
                if(b_new==0):
                    state_new1=[(state[0]-i),(state[1]-j),b_new]
                    if(validState(state_new1) and state_new1!=state):
                        childstates.append(state_new1)                     
                elif(b_new==1):
                    state_new2=[(state[0]+i),(state[1]+j),b_new]
                    if(validState(state_new2) and state_new2!=state):                            
                        childstates.append(state_new2)
    return childstates
        
    
def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        s = queue.pop(0)
        
        children=nextStates(s)
        for neighbour in children:
            while(neighbour not in visited):
                visited.append(neighbour)
                queue.append(neighbour)
                
                #saves the parent value of child to path
                path.append([neighbour,s])
                
                if(neighbour==goal):
                    return path

path=bfs(visited, states, start)
print(path)
```

```python
#from the path, we need to find the shortest path to the goal state

#set child as goal node
child=path[-1][0]
parent=path[-1][1]

result=[]
result.append(child)
result.append(parent)

#while parent not 'start' node
while(parent!=[3,3,1]):
    
    #find parent of parent
    for index, sublist in enumerate(path):    
        
        #if sublist contains the parent in child position  
        if sublist[0] == parent:
            
            #assign the new parent and child
            parent=sublist[1]
            child=sublist[0]
            result.append(sublist[1])
result.reverse()
print(result)
```

```python
import pandas as pd 
df = pd.DataFrame(result)
df=df.rename(columns={0:'Missionaries',1:'Cannibals',2:'Boat'})
df['Missionaries-Side2']=3-df['Missionaries']
df['Cannibals-Side2']=3-df['Cannibals']
df['Boat'] = df['Boat'].replace(1, 'Start Side')
df['Boat'] = df['Boat'].replace(0, 'Goal Side')
df.style.hide_index()
```
